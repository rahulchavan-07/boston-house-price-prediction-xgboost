Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the Boston House price Dataset

```python
house_price_dataset=pd.read_csv('boston.csv')
```

```python
print(house_price_dataset)
```

```python
#print first five rows
house_price_dataset.head()
```

```python
house_price_dataset['MEDV']
```

```python
#check number of rows and column
house_price_dataset.shape
```

```python
#check for missing values
house_price_dataset.isnull().sum()
```

```python
#staticstical message of the dataset
house_price_dataset.describe()
```

Understanding the corelation between various feature in dataset

positive corelation
negative corelation

```python
correlation=house_price_dataset.corr()
```

```python
#constructing heatmap to undestand the corelation

plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':'8'},cmap='Blues')
```

Splitting the data and target

```python
X=house_price_dataset.drop(['MEDV'],axis=1)
Y=house_price_dataset['MEDV']
```

```python
print(X)
print(Y)
```

Splitting the data into training

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

model training (XGBoost Regressor)

```python
#loading the model
model=XGBRegressor()
```

```python
# training the model with X_train

model.fit(X_train,Y_train )
```

Evaluation

```python
#accuracy for prediction on training data
training_data_prediction=model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error

score_1 = metrics.r2_score(Y_train,training_data_prediction)
score_2=metrics.mean_absolute_error(Y_train,training_data_prediction)
print("R squared error : " , score_1)
print("mean absolute error : " , score_2)
```

 visualizing the actual price and predicted price


```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("actual prices")
plt.ylabel("predicted prices")
plt.title("actual price vs predicted prices")
plt.show()
```

Prediction on test data

```python
test_data_prediction=model.predict(X_test)
```

```python
score_1 = metrics.r2_score(Y_test,test_data_prediction)
score_2=metrics.mean_absolute_error(Y_test,test_data_prediction)
print("R squared error : " , score_1)
print("mean absolute error : " , score_2)
```

 visualizing the actual price and predicted price


```python
plt.scatter(Y_test,test_data_prediction)
plt.xlabel("actual prices")
plt.ylabel("predicted prices")
plt.title("actual price vs predicted prices")
plt.show()
```
